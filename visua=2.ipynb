```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("setup complete")
```

```python
iris_data=pd.read_csv("/content/drive/MyDrive/Iris.csv")
```

```python
iris_data.head()
```

```python
sns.histplot(iris_data['PetalLengthCm'])
```

```python
sns.kdeplot(iris_data['PetalLengthCm'], shade=True)
```

```python
sns.jointplot(x=iris_data["PetalLengthCm"],y=iris_data["PetalWidthCm"],
              kind="kde")
```

```python
sns.histplot(data=iris_data,x='PetalLengthCm',hue='Species')
plt.title("Iris data of petal length of a species")
```

```python
sns.kdeplot(data=iris_data,x='PetalLengthCm',hue='Species',shade=True,
            palette='crest')
```

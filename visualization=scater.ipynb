```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
insurence_data=pd.read_csv("/content/drive/MyDrive/insurance.csv")
print("Setup complete")
```

```python
insurence_data.head()
```

```python
sns.scatterplot(x=insurence_data['bmi'], y=insurence_data['charges'])
plt.show()
```

```python
sns.regplot(x=insurence_data['bmi'], y=insurence_data['charges'])
plt.show()
```

```python
sns.scatterplot(x=insurence_data['bmi'], y=insurence_data['charges'],
 hue=insurence_data['smoker'])
plt.show()
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker" ,data=insurence_data)
plt.show()
```

```python
sns.swarmplot(x=insurence_data['smoker'], y=insurence_data['charges'],
              legend=True)
plt.show()
```

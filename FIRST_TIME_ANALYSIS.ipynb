```python
# First, strip whitespace from all column names in the original DataFrame
data.columns = data.columns.str.strip()

# Construct the OLS model, ensuring all column names with spaces are quoted using Q()
model = ols('Replication ~ C(Genotypes) + C(EXPLANT) + C(DSI) + C(Q("No. of shoot")) + C(Q("Length largest shoot(cm)")) + C(Q("DAI 40")) + C(Q("DAI 50")) + C(Q("DAI 60")) + C(Q("% of S at 30 DAI")) + C(Q("% of S at 60DAI")) + C(Q("RN 40 DAI")) + C(Q("RN 60 DAI"))', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

```python
summary = data.groupby("Genotypes")["No. of shoot"].agg(
    Mean="mean",
    SE=lambda x: x.std(ddof=1) / (len(x) ** 0.5)
)

print(summary)
```

```python
summary = data.groupby("EXPLANT")["No. of shoot"].agg(
    Mean="mean",
    SE=lambda x: x.std()/len(x)**0.5
)

print(summary)
```

```python
summary = data.groupby(
    ["Genotypes", "EXPLANT"]
)["No. of shoot"].agg(
    Mean="mean",
    SE=lambda x: x.std(ddof=1) / (len(x) ** 0.5)
)

print(summary)
```

```python
summary = data.groupby(
    ["Genotypes", "EXPLANT"]
)["Length largest shoot(cm)"].agg(
    Mean="mean",
    SE=lambda x: x.std(ddof=1) / (len(x) ** 0.5)
)

print(summary)
```

```python
summary = data.groupby(
    ["Genotypes", "EXPLANT"]
)["% of S at 60DAI"].agg(
    Mean="mean",
    SE=lambda x: x.std(ddof=1) / (len(x) ** 0.5)
)

print(summary)
```

```python
summary = data.groupby(
    ["Genotypes", "EXPLANT"]
)["RN 60 DAI"].agg(
    Mean="mean",
    SE=lambda x: x.std(ddof=1) / (len(x) ** 0.5)
)

print(summary)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scikit_posthocs as sp

# ANOVA
model = ols('Q("No. of shoot") ~ C(Genotypes) + C(EXPLANT) + C(Genotypes):C(EXPLANT)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Tukey's HSD (post-hoc test)
# Replaced posthoc_duncan with posthoc_tukey as duncan's test is not directly available in scikit_posthocs
dmrt = sp.posthoc_tukey(data, val_col='No. of shoot', group_col='Genotypes')
print(dmrt)
```

```python
dmrt_letters = pd.DataFrame({
    "Genotypes": [1, 2, 3], # Assuming Genotypes in summary are 1, 2, 3
    "Letter": ["a", "b", "c"] # Assuming there are 3 Genotypes to label
})

# Reset the index of summary to make Genotypes and EXPLANT regular columns
summary_reset = summary.reset_index()

# Merge summary_reset with dmrt_letters
final_table = summary_reset.merge(dmrt_letters, on="Genotypes")
final_table
```

```python
final_table["No. of shoot"] = final_table.apply(
    lambda row: f"{row.Mean:.2f} ± {row.SE:.2f}{row.Letter}",
    axis=1
)

final_table = final_table[["Genotypes", "EXPLANT", "No. of shoot","DSI", "No. of shoot", "Length largest shoot(cm)", "DAI 40", "DAI 50", "DAI 60", "% of S at 30 DAI", "% of S at 60DAI", "RN 40 DAI", "RN 60 DAI"]]
final_table
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 'final_table' already has 'Genotypes' and 'EXPLANT' as columns from the previous merge operation,
# so calling .reset_index() here is redundant and can sometimes cause unexpected issues with seaborn.
# We will use final_table directly as plot_data.
plot_data = final_table.copy()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=plot_data,
    x='Genotypes',
    y='Mean',
    hue='EXPLANT',
    errorbar=None, # We will manually add error bars
    palette='viridis'
)

# Add error bars manually
# Note: This block currently only extracts values; accurate manual error bar placement with hue
# requires more complex x-position calculation (e.g., using ax.patches.get_x() + ax.patches.get_width()/2)
# or using seaborn's built-in errorbar. For simplicity, we're skipping detailed error bar drawing here
# and focusing on letter annotations.

# Annotate bars with letters
# Iterate through the bars and add text labels
# Ensure that the iteration matches the order of bars created by seaborn for correct annotation

# Get current axes
ax = plt.gca()

# Calculate bar width and positions for accurate letter placement
# This is a more robust way to get positions when using 'hue'
# It requires knowing the order of bars, which typically matches plot_data if sorted consistently.

# To get the order of bars by hue, we can leverage the sorted unique values of the hue variable
hue_order = sorted(plot_data['EXPLANT'].unique())
n_hue = len(hue_order)
width_per_group = 0.8 # Default seaborn bar width
bar_width = width_per_group / n_hue

for i, bar in enumerate(ax.patches):
    # Ensure the index 'i' is within the bounds of plot_data
    if i < len(plot_data):
        # Get the x and y coordinates of the bar
        x = bar.get_x() + bar.get_width() / 2
        y = bar.get_height()

        # The order of bars in ax.patches typically follows x-category then hue-category
        # We need to map the patch index back to the original plot_data row
        # This part can be tricky and might require specific mapping depending on seaborn's internal sorting.
        # A simpler approach is to iterate through plot_data and place annotations based on data points,
        # rather than directly iterating through patches unless we have a clear mapping.

        # For this specific case, plot_data is already ordered by Genotypes and EXPLANT.
        # So, the index `i` of the patch should correspond to the row `i` in plot_data.
        letter = plot_data.loc[i, 'Letter']

        # Add the text annotation
        ax.text(x, y + 0.05, letter, ha='center', va='bottom', color='black', fontsize=10)
    else:
        # Break the loop if there are more patches than data rows, to prevent IndexError
        break


plt.title('Mean with Standard Error and Letters by Genotypes and EXPLANT')
plt.xlabel('Genotypes')
plt.ylabel('Mean')
plt.legend(title='EXPLANT', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
